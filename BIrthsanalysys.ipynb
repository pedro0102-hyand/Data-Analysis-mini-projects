```python
import numpy as np
import pandas as pd
```

```python
birth=pd.read_csv("births.csv")
birth.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
sns.set()
```

```python
quartiles=np.percentile(birth["births"],[25,50,75])
mean,sigma=quartiles[1],0.74*(quartiles[2]-quartiles[0])
birth=birth.query("(births>@mean -5 * @sigma)&(births<@mean + 5 * @sigma)")
birth["day"]=birth["day"].astype(int)
birth.index=pd.to_datetime(10000*birth.year+100*birth.month+birth.day,format="%Y%m%d")
birthsDate=birth.pivot_table("births", [birth.index.month,birth.index.day])
fig,ax=plt.subplots(figsize=(12,4))
birthsDate.plot(ax=ax)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,4))
birthsDate.plot(ax=ax)
style=dict(size=10,color='black')
ax.text(pd.to_datetime("2012-1-1"),3950,"New Year",**style)
ax.text(pd.to_datetime("2012-7-4"),4250,"Independence Day",ha="center",**style)
ax.text(pd.to_datetime("2012-9-4"),4850,"Labour Day",ha="center",**style)
ax.text(pd.to_datetime("2012-10-31"),4600,"Halloween",ha="right",**style)
ax.text(pd.to_datetime("2012-12-25"),4450,"Thanksgiving",ha="center",**style)
ax.text(pd.to_datetime("2012-12-25"),3850,"Christmas",ha="right",**style)
ax.set(title="USA births per day",ylabel="average daily births")
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter("%h"))
plt.show()
```

```python
fig,ax=plt.subplots(facecolor='lightgrey')
ax.axis([0,10,0,10])
ax.text(1,5,".Data:(1,5)",transform=ax.transData)
ax.text(0.5,0.1,".Axes:(0.5,0.1)",transform=ax.transAxes)
ax.text(0.2,0.2,".Figure:(0.2,0.2)",transform=fig.transFigure)
plt.show()
```
